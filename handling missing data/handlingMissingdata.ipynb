# Handling missing data

### dropna() method
-   dropna() -> used to drop rows or columns with missing data.

-   axis = 0 means remove row
-   axis = 1 means remove columns

```python
import pandas as pd
```

```python
data = {
    "Name": ["Ram", None, "Ghansyam", "Kam", "Aaram", "Sharam", "Gharam"],
    "Age": [12, 13, 14, 15, 16, 17, 18],
    "City": ["Delhi", "Mumbai", "Kolkata", None, "Bangalore", "Hyderabad", "Pune"],
    "Salary": [1000, 2000, 3000, None, 5000, None, 7000],
    "Performance Score": [56, 85, 95, 62, 75, 88, 90],
}

df = pd.DataFrame(data)
df
```

```python
df.dropna(axis=0, inplace=True)
df
```

```python
df.dropna(axis=1, inplace=True)
df
```

-   Remove row or column which ever unneeded.

### fillna() method
-   fillna() -> used to fill missing values with a specific value.

```python
df_fill = pd.DataFrame(data)
df_fill
```

```python
df_fill.fillna(0, inplace=True)
df_fill
```

```python
df_2 = pd.DataFrame(data)
df_2['Salary'].fillna(df_2['Salary'].mean(), inplace=True)
df_2
```
