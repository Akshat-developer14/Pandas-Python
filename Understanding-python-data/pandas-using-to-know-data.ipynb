## Problem with data is we don't know
-   1. How many columns and rows?
-   2. What type of data it is?
-   3. Missing data?

### To solve we can use different methods like
-   1. head()
-   2. tail()
-   3. info()
-   4. describe()
-   5. shape
-   6. isnull()
-   7. notnull()
-   8. duplicated()

### What problem they solve
-   1. How many columns and rows? - shape
-   2. What type of data it is? - info()
-   3. Missing data? - isnull()
-   4. Columns name.
-   5. Non null counts.
-   6. Memory usage of the dataframe.
